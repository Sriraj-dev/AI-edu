# Matplotlib

```python
import pandas as pd 
import numpy as np

%matplotlib inline
from matplotlib import style
style.use('default')
df = pd.DataFrame(np.random.randn(100,4),columns=["Mon","Tue","Wed","Thu"])
```

## Histograms

```python
# histogram
df.hist()
```

```python
df['Wed'].hist(bins=50)
```

```python
#another way
df['Wed'].plot(kind='hist', bins = 60)
```

```python
df['Wed'].hist(bins=50,legend = True,grid=False)
```

## Area plots

```python
#in area plots we need all the colums to be either all +ve or all -ve
# so we set stacked = false, to allow it.

#first 25 rows
df.iloc[:26].plot(kind='area',stacked=False)
```

```python
# shading
df.iloc[:26].plot(kind='area',stacked=False, alpha = 0.75)
```

```python
# title
df.iloc[:26].plot(kind='area',stacked=False, title="Area plt")
```

```python
# legend
df.iloc[:26].plot(kind='area',stacked=False,legend= False)
```

```python
# plot only one column
df.iloc[:26]['Thu'].plot(kind='area',stacked=False,legend=True)
```

```python
#adding table
df.iloc[:26].plot(kind='area',stacked=False,table=True)
```

## Bar graphs

```python
df.iloc[:12].plot(kind='bar')
```

```python
# bar plot stacked
# converting the data to +ve for clear clarity.
df.iloc[:20].abs().plot(kind = 'bar',stacked=True)
```

```python
#shading / grid / Title / Legend  -> you can modify
df.iloc[:20].abs().plot(kind = 'bar',stacked=True,grid= False, title='BarGraph')
```

```python
#second way
df.iloc[:26].abs().plot.bar(stacked = True)
# again, you can just change all the parameters
```

## Line Charts

```python
df.iloc[:20].plot(kind='line')
```

```python
# Absolute line plot
df.iloc[:20].abs().plot(kind='line')
```

```python
# line width
df.iloc[:20].plot(kind='line',lw = 5)
```

```python
# column
df.iloc[:20]['Mon'].plot(kind='line')
```

```python
# legend
df.iloc[:20].plot(kind='line',legend=True)
```

```python
# size
df.iloc[:20].plot(kind='line',figsize=(10,5))
```

```python
# Title / Shading / Legend etc..
df.iloc[:20].plot(kind='line',title="Line chart", alpha = 0.5)
```

## Scatterplots

```python
# scatter plots
df.plot(kind='scatter',x='Mon',y='Tue')
```

```python
# color sequence:
df.plot(kind='scatter',x='Mon',y='Tue',c= 'Wed',legend=True)
```

```python
# cmap colors
df.plot(kind='scatter',x='Mon',y='Tue',c= 'Wed',colormap='magma')
```

```python
#size
df.abs().plot(kind='scatter',x='Mon',y='Tue',s= 'Wed')
```

```python
# for clarity
df.plot(kind='scatter',x='Mon',y='Tue',s = df['Wed'].abs()*50)
```

```python
# add alpha for better visualisation
df.plot(kind='scatter',x='Mon',y='Tue',s = df['Wed'].abs()*50,alpha = 0.5)
```

## Box plots

```python
df.plot(kind='box')
```

```python
# size
df.plot(kind='box',figsize=(10,5))
```

```python
#color
df.plot(kind='box',colormap='jet')
```

## Hex Bin plots

```python
#hex bin
df.plot(kind='hexbin',x='Mon',y='Tue')
```

```python
# grid size
df.plot(kind='hexbin',x='Mon',y='Tue',gridsize = 10)
```

```python
#color
df.plot(kind='hexbin',x='Mon',y='Tue',gridsize = 10,colormap='Greens')
```

```python
df.plot.hexbin(x='Mon',y='Tue',gridsize = 10,colormap='Greens')
```

## Density & KDE plots

```python
df['Mon'].plot(kind='kde')
```

```python
# density plot - same
df['Mon'].plot(kind='density')
```

```python
df.plot.density(lw = 2,grid = True, title = "Density curve")
```
